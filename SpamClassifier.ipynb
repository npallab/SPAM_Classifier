```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

Importing the DataSet

```python
df=pd.read_excel('/content/spam.xls')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.rename(columns={'v1':'target','v2':'text'},inplace=True)
```

```python
df.head()
```

```python
df['target']=np.where(df['target']=='spam',1,0)
```

```python
df.head()
```

```python
df['target'].value_counts()
```

Upsampling

```python
df_maj=df[df['target']==0]
df_min=df[df['target']==1]
```

```python
from sklearn.utils import resample
```

```python
df_resampled=resample(df_min,n_samples=len(df_maj),random_state=123)
```

```python
df_fin=pd.concat([df_maj,df_resampled])
```

```python
df_fin['target'].value_counts()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vector=TfidfVectorizer()
target=df_fin['target']
feature=df_fin['text'].astype('str')
vector.fit(feature) # Fit the vectorizer to your text data
feature_transformed = vector.transform(feature).toarray() # Transform your text data and convert to a dense array
# Now 'feature_transformed' contains the numerical representation of your text data, and 'feature' still holds your original text data
```

```python
feature_transformed=pd.DataFrame(feature_transformed,columns=vector.get_feature_names_out())
```

```python
feature_transformed
```

```python
from sklearn.model_selection import train_test_split # Import train_test_split from the correct module: model_selection
x_train,x_test,y_train,y_test=train_test_split(feature_transformed,target,test_size=0.2,random_state=123)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier=RandomForestClassifier()
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
accuracy=accuracy_score(y_test,y_pred)
conf_met=confusion_matrix(y_test,y_pred)
class_rep=classification_report(y_test,y_pred)
```

```python
print(accuracy)
print(conf_met)
print(class_rep)
```

```python
import gradio as gr
import pickle as pkl

# Load the model and vectorizer
classifier = pkl.load(open('spam_class.pkl', 'rb'))
vectorizer = pkl.load(open('vector_spam.pkl', 'rb'))

def predict_spam(input_text):
    # Transform the input text using the vectorizer
    transformed_text = vectorizer.transform([input_text])
    # Predict using the classifier
    prediction = classifier.predict(transformed_text)
    # Return result
    if prediction[0] == 0:
        return "Not Spam"
    else:
        return "Spam"

# Define the Gradio interface
interface = gr.Interface(
    fn=predict_spam,
    inputs=gr.Textbox(lines=2, placeholder="Enter text here...", label="Text Input"),
    outputs=gr.Textbox(label="Prediction"),
    title="Spam Detection App",
    description="Enter a text message to predict if it's Spam or Not Spam."
)

# Launch the interface
interface.launch()
```
